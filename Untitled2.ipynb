**UVA:573 The Snail**

Có 1 con ốc sên nằm ở dưới đáy của 1 cái giếng có chiều cao là H (m).Ốc sên có thể  bò lên U(m) khi trời sáng, nhưng sẽ bị tuột xuống D(m) vào buổi tối khi đi ngủ. Sau mỗi ngày, ốc sên sẽ bò ít hơn ngày đầu tiên F%, hỏi  vào ngày thứ bao nhiêu thì ốc sên sẽ bò ra khỏi miệng giếng hoặc là rớt lại xuống đáy giếng.

**Input:**

File đầu vào bao gồm 1 hoặc nhiều test cases, mỗi test case sẽ nằm trên 1 dòng, bao gồm 4 số nguyên H,U,D,F cách nhau bởi khoảng trắng. Nếu H = 0, tức là kết thúc số test case. 

Lưu ý: Ốc sên sẽ không bao giờ bò 1 khoảng cách âm.Nếu ngày hôm đó ốc sên bò U < 0 thì tức là hôm đó nó đứng yên không di chuyển và chỉ bị tuột xuống vào buổi tối. 

**Output:**

Với mỗi test case, xuất ra một dòng thông báo liệu ốc sên có thành công (ra khỏi miệng giếng) hay là thất bại (rớt xuống đáy).

**Sample input:**

6 3 1 10

10 2 1 50

50 5 3 14

1 1 1 1

0 0 0 0

**Output**

success on day 3

failure on day 4

failure on day 7

failure on day 2

**Lời giải :**

Chúng ta tính lại khoảng cách của con ốc sên với đáy giếng sau mỗi ngày, nếu khoảng cách lớn hơn chiều cao của giếng thì tức là con ốc sên thoát được khỏi miệng giếng, còn nếu khoảng cách bé hơn hoặc bằng không thì con ốc sên bị kẹt lại ở đáy giếng. 




```python
while True:
	H , U ,D ,F = list(map(int,input().split()))

	if H == 0:
		break

	day = 1
	distance = 0
	u = F*U/100

	while True :
		distance += U
		if U > 0:
			U -= u
		if U < 0:
			U = 0
		if distance > H:
			break
		else:
			distance -= D
		if distance < 0:
			break
		else:
			day += 1

	if(distance < 0):
		print("failure on day {}".format(day))
	else:
		print("success on day {}".format(day))
```

```python
#include <bits/stdc++.h>

int main()
{
	int H, D, F, dia;
	float U, alt, f;

	while(1)
	{
		scanf("%d %f %d %d", &H, &U, &D, &F);
		if(H == 0) break;
		dia = 1;		
		alt = 0;
		f = U*F/100;
		while(1){	
			alt += U;
			if(U > 0)
				U -= f;
			if(alt > H) break;			
			alt -= D;			
			if(alt < 0) break;			
			dia++;
		}
		if(alt < 0)
			printf("failure on day %d\n", dia);
		else
			printf("success on day %d\n", dia);
	} 	
	return 0;
}
```

**UVA 10141: Request for proposal**

Khi chính phủ có nhu cầu thực hiện một giao dịch mua  lớn.
Đầu tiên họ sẽ phát hành 1 danh sách  liệt kê các yêu cầu cần thiết
và các nhà cung ứng sẽ đề xuất danh sách những yêu cầu họ đáp ứng được kèm 
với mức giá tương ứng.
Việc của bạn là chọn ra nhà cung ứng đáp ứng nhiều nhất số yêu cầu trong danh sách 
 cùng với mức giá thấp nhất.

 việc đáp ứng yêu cầu được tính như sau:

    điểm = số yêu cầu đáp ứng được / tổng số yêu cầu có trong danh sách. 

**lưu ý**: nếu có 2 công ty có cùng số điểm thì công ty có mức giá thấp hơn sẽ được chọn,
nếu cả 2 công ty đều cùng điểm và cùng mức giá thì công ty nào đứng trước sẽ được chọn.

**Input:**
- dòng đầu bao gồm 2 số : 0 < n <= 1000 : số yêu cầu trong danh sách,  p: số nhà cung ứng.
Lưu ý là có thể có nhiều danh sách yêu cầu do đó dòng 0 0 sẽ kết thúc quá trình nhập danh sách.
- n dòng tiếp theo sẽ bao gồm các yêu cầu trong danh sách.
- trong mỗi p nhà cung ứng:
    - 1 dòng là tên nhà cung ứng 
    - 1 dòng bao gồm d(float): mức giá và r:( 0 <= r <= n):số yêu cầu đáp ứng được.
    - r dòng in ra yêu cầu đáp ứng được trên mỗi dòng.

**Output:**
- In ra tên nhà cung ứng tốt nhất, để 1 dòng trống giữa các nhà cung ứng.

**Lời giải:**

Để giải bài này chúng ta nên liên tục cập nhật 2 giá trị  best_score và best_cost.
Nếu 1 nhà cung ứng có score > best_score hoặc score = best_score nhưng cost < best_cost 
thì chúng ta sẽ lưu nhà cung ứng dó lại.Lưu ý là bạn không cần phải lưu lại tên các yêu cầu 
trong danh sách, chúng ta chỉ cần quan tâm đến r và n để tính lại best_score.

```python
##include<bits/stdc++.h>
 
using namespace std;
 
int n, p, r;
float d, best_compliance, best_cost;
string line, name, best_name;
 
int main() {
    for(int T = 1;; T++) {
        scanf("%d %d\n", &n, &p);
        if(n == 0 && p == 0) break;
        if(T > 1) cout << endl;
        for(int i = 0; i < n; i++)
            getline(cin, line);
        best_compliance = 0.0;
        best_cost = 1E37;
        for(int i = 0; i < p; i++) {
            getline(cin, name);
            scanf("%f %d\n", &d, &r);
            for(int i = 0; i < r; i++)
                getline(cin, line);
            if((float) r / p > best_compliance \
                    || (float) r / p == best_compliance && d < best_cost) {
                best_compliance = (float) r / p;
                best_cost = d;
                best_name = name;
            }
        }
        cout << "RFP #" << T << endl;
        cout << best_name << endl;
    }
}
```
